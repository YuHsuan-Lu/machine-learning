## 1. Decision Trees as Interpretable Models

Import packages

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import math
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

### (a) Obtain Data

Get the Accute Inamations Data Set

```python
columns = ['Temperature of patient', 'Occurrence of nausea', 'Lumbar pain', 'Urine pushing', 'Micturition pains',
       'Burning of urethra, itch, swelling of urethra outlet ', 'decision: Inflammation of urinary bladder',
       'decision: Nephritis of renal pelvis origin']

df = pd.read_csv('data/diagnosis.data.csv', header=None, encoding="utf-16", sep='\t', names=columns)
df.head()
```

```python
df = df.replace(to_replace=['no', 'yes'], value=[0, 1])
df.head()
```

```python
df['Temperature of patient'] = pd.to_numeric(df['Temperature of patient'].str.replace(',', '.'))
df.head()
```

### (b) Build a decision tree

```python
from sklearn.multioutput import MultiOutputClassifier
```

```python
y1 = df['decision: Inflammation of urinary bladder']
y2 = df['decision: Nephritis of renal pelvis origin']

X = df.drop(['decision: Inflammation of urinary bladder', 'decision: Nephritis of renal pelvis origin'], axis=1)

Y = np.vstack((y1, y2)).T
```

```python
# dt = MultiOutputClassifier(DecisionTreeClassifier(criterion='entropy', random_state=0), n_jobs=2)
dt = DecisionTreeClassifier(criterion='entropy', random_state=0)
dt.fit(X, Y)
```

```python
fig = plt.figure(figsize=(25,20))
tree.plot_tree(dt, feature_names=list(X.columns), class_names=[['no', 'yes'], ['no', 'yes']])
plt.show()
```

### (c) Convert the decision rules

```python
from sklearn.tree import _tree

def tree_to_code(tree, feature_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]
    feature_names = [f.replace(" ", "_")[:-5] for f in feature_names]
    print("def predict({}):".format(", ".join(feature_names)))

    def recurse(node, depth):
        indent = "    " * depth
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            print("{}if {} <= {}:".format(indent, name, np.round(threshold,2)))
            recurse(tree_.children_left[node], depth + 1)
            print("{}else:  # if {} > {}".format(indent, name, np.round(threshold,2)))
            recurse(tree_.children_right[node], depth + 1)
        else:
            print("{}return {}".format(indent, tree_.value[node]))

    recurse(0, 1)
tree_to_code(dt, feature_names=list(X.columns))
```

### (d) Cost-Complexity Pruning

```python
path = dt.cost_complexity_pruning_path(X, Y)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
fig, ax = plt.subplots()
ax.plot(ccp_alphas[:-1], impurities[:-1], marker="o", drawstyle="steps-post")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
plt.show()
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(X, Y)
    clfs.append(clf)
print(
    "Number of nodes in the last tree is: {} with ccp_alpha: {}".format(
        clfs[-1].tree_.node_count, ccp_alphas[-1]
    )
)
```

```python
train_scores = [clf.score(X, Y) for clf in clfs]
test_scores = [clf.score(X, Y) for clf in clfs]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker="o", label="train", drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker="o", label="test", drawstyle="steps-post")
ax.legend()
plt.show()
```

## 2. The LASSO and Boosting for Regression

### (a) Obtain Data

```python
df2 = pd.read_csv("data/communities.data.csv", header=None, na_values=['?'])
df2.head()
```

```python
df2 = df2.iloc[: , 5:]
df2.head()
```

```python
df2.shape
```

```python
df_train = df2.iloc[0:1495,]
df_test= df2.iloc[1495:,]
```

### (b) Missing values

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(strategy='mean')
imp_mean.fit(df_train)

df2 = pd.DataFrame(imp_mean.transform(df2))
df_train = pd.DataFrame(imp_mean.transform(df_train))
df_test = pd.DataFrame(imp_mean.transform(df_test))
```

### (c) Plot a correlation matrix

```python
plt.figure(figsize=(25, 10))
sns.heatmap(df_train.corr())
plt.title('Correlation Matrix')
plt.show()
```

### (d) Calculate the Coefficient of Variation CV

```python
cv = np.apply_along_axis(lambda x: np.var(x) / np.mean(x), axis=0, arr=df2.values)
cv
```

### (e) Scatter plots and box plots for highest CV features

```python
cols = df2.columns.values
col_cv = pd.DataFrame(np.column_stack([cols, cv]))
col_cv = col_cv.drop(0)
col_cv = col_cv.sort_values(1, ascending=False)
print("Coeficients of all the coloumns")
print(col_cv)

high_var11 = col_cv.head(int(math.floor(math.sqrt(128))))
selected_cols = df2[high_var11[0]]
print("Selected 11 Coloumns ")
print(high_var11)
```

```python
sns.pairplot(selected_cols)
```

Some features move together in a linear way in other words  are linearly correlated. For other features, we cannot say anything meaningful about their relationship using above figures.

```python
sns.boxplot(data=selected_cols)
plt.title('Box plot of Selected Features')
plt.show()
```

### (f) Fit a linear model

```python
X_train = df_train.iloc[:, :-1].values
X_test = df_test.iloc[:, :-1].values
y_train = df_train.iloc[:, -1].values 
y_test = df_test.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

lr = LinearRegression().fit(X_train, y_train)
y_pred = lr.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)

print('Mean Absolute Error: %.3f' % mae)
```

### (g) Fit a ridge regression model

```python
from sklearn.linear_model import RidgeCV, Ridge
```

```python
ridge = RidgeCV(alphas=[1e-3, 1e-2, 1e-1,0.06,0.05, 1]).fit(X_train, y_train)
print("Best Lambda:", ridge.alpha_)
```

```python
ridge = Ridge(alpha=1.0).fit(X_train, y_train)
y_pred = ridge.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)

print('Mean Absolute Error: %.3f' % mae)
```

### (h) Fit a LASSO model

```python
from sklearn.linear_model import LassoCV, Lasso
```

```python
lasso = LassoCV(alphas=[1e-3, 1e-2, 1e-1,0.06,0.05, 1]).fit(X_train, y_train)
print("Best Lambda:", lasso.alpha_)
```

```python
lasso = Lasso(alpha=0.001).fit(X_train, y_train)
y_pred = lasso.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)

print('Mean Absolute Error without Standardization: %.3f' % mae)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lasso = LassoCV(alphas=[1e-3, 1e-2, 1e-1,0.06,0.05, 1]).fit(X_train_scaled, y_train)
print("Best Lambda:", lasso.alpha_)
```

```python
lasso = Lasso(alpha=0.001).fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)

mae = mean_absolute_error(y_test, y_pred)

print('Mean Absolute Error with Standardization: %.3f' % mae)
```

### (i) Fit a PCR model

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold
```

```python
maes = []

for n in range(1, X_train.shape[1]+1):
    pca = PCA(n)
    X_train_PCA = pca.fit_transform(X_train)
    mae = []
    for train_index, val_index in KFold(n_splits=10).split(X_train_PCA):
        x_train, x_val = X_train_PCA[train_index.tolist()], X_train_PCA[val_index.tolist()]
        yy_train, y_val = y_train[train_index.tolist()], y_train[val_index.tolist()]
        
        lr = LinearRegression().fit(x_train, yy_train)
        y_pred = lr.predict(x_val)
        mae.append(mean_absolute_error(y_val, y_pred))
    maes.append(np.mean(mae))
```

```python
sns.lineplot(x=range(1, X_train.shape[1]+1), y=maes)
plt.xlabel('Number of Components')
plt.ylabel('CV MAE')
plt.show()
```

```python
pd.DataFrame({'n': range(1, X_train.shape[1]+1), "CV MAE": maes}).sort_values('CV MAE', ascending=True)
```

```python
pca = PCA(65)
X_PCA_train = pca.fit_transform(X_train)
X_PCA_test = pca.fit_transform(X_test)

lr = LinearRegression().fit(X_PCA_train, y_train)
y_pred = lr.predict(X_PCA_test)

mae = mean_absolute_error(y_test, y_pred)

print('Mean Absolute Error: %.3f' % mae)
```

### (j) Fit a boosting tree

```python
!pip install xgboost
from xgboost import XGBRegressor
```

```python
maes = []

alphas = [0.01, 0.05, 0.1, 0.2, 0.5, 0.75, 1.2, 1.4, 1.6, 1.8, 2, 3, 4, 5, 10, 20]
for alpha in alphas:
    mae = []
    for train_index, val_index in KFold(n_splits=10).split(X_train):
        x_train, x_val = X_train[train_index.tolist()], X_train[val_index.tolist()]
        yy_train, y_val = y_train[train_index.tolist()], y_train[val_index.tolist()]
        
        xgb = XGBRegressor(random_state=0, reg_alpha=alpha).fit(x_train, yy_train)
        y_pred = xgb.predict(x_val)
        mae.append(mean_absolute_error(y_val, y_pred))
    maes.append(np.mean(mae))
```

```python
sns.lineplot(x=alphas, y=maes)
plt.xlabel('alpha')
plt.ylabel('CV MAE')
plt.show()
```

```python
pd.DataFrame({'alpha': alphas, "CV MAE": maes}).sort_values('CV MAE', ascending=True)
```

```python
xgb = XGBRegressor(random_state=0, reg_alpha=1.60).fit(X_train, y_train)
y_pred = xgb.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)

print('Mean Absolute Error: %.3f' % mae)
```
